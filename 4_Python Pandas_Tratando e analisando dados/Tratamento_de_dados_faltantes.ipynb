## Tratamento de dados faltantes

```python
import pandas as pd
```

```python
dados = pd.read_csv('aluguel_residencial.csv', sep=';')
```

```python
dados.head(20)
```

## Função isnull

```python
dados.isnull() # Para visualizar os dados faltantes. Ele marca True nos dados nulos.
```

## Função notnull

```python
dados.notnull() #O mesmo acontece com o notnull, só que o contrário. Marca True em todos os dados que não são nulos. 
```

```python
dados.info()
```

```python
dados['Valor'].isnull() # Cria uma series boleana
```

```python
dados[ dados['Valor'].isnull()] #Aqui, porque foi inserido um par de [] a mais ele já apresenta os dados na estrutura de dataframe
#Aqui temos todos os registros com 'Valor' de dados nulos. 
```

```python
A = dados.shape[0] # verificaremos o tamanho do dataframe antes de eliminar os valores nulos.
```

## Função dropna
tem o objetivo de eliminar os dados nulos.

```python
dados.dropna(subset = [ 'Valor'], inplace = True)
B = dados.shape[0]
A - B
```

```python
dados[ dados['Valor'].isnull()] # Como na linha superior foi eliminado todos os dados nulos de "Valor" faz sentido que agora nenhum dado seja apresentado nesta tabela. 
```

## Tratamento de Dados de Condomínio Faltantes

```python
dados[dados['Condominio'].isnull()].shape[0]
```

```python
selecao = (dados['Tipo']== 'Apartamento') & (dados[ 'Condominio'].isnull())
```

```python
A = dados.shape[0]
dados = dados[~selecao] # O ~ inverte a series boleana, o que é False fica True e o que é True fica False
B = dados.shape[0]
A - B
```

```python
dados[dados['Condominio'].isnull()].shape[0]
```

##Função fillna(n)
Substitui todos os valores do nulo para o número definido na função, que no caso aqui foi o zero. 

```python
dados = dados.fillna({'Condominio':0, 'IPTU': 0}) # Foi usado um dicionário pois assim é possível alterar o valor nulo de diferentes variáveis.
```

```python
dados[dados['Condominio'].isnull()].shape[0] # Verificando a quantidade de dados nulos. 
```

```python
dados[dados['IPTU'].isnull()].shape[0] # Verificando a quantidade de dados nulos. 
```

```python
dados.info()
```

```python
dados.to_csv('aluguel_residencial.csv', sep=';', index = False)
```
