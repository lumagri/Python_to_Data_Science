#Relatório de Análises VI
##Criando novas variáveis

```python
import pandas as pd
```

```python
dados = pd.read_csv('aluguel_residencial.csv', sep = ';')
```

```python
dados.head(10)
```

```python
#Criar variáveis de valor bruto

dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

```python
dados.head(10)
```

```python
#Criar variável de m2

dados[ 'Valor m2'] = dados['Valor'] / dados['Area']
```

```python
dados.head(10) # Fica feio muitas casas depois da virgula. Podemos arredondar o valor na tabela pra ficar melhor estéticamente. 
```

```python
dados['Valor m2'] = dados['Valor m2'].round(2) #Apresenta apenas 2 casas depois da vírgula devido ao 2 dentro da função round.
```

```python
dados.head(10)
```

```python
#Criar variável de Valor Bruto por m2

dados[ 'Valor Bruto m2'] = (dados['Valor Bruto'] / dados['Area']).round(2)
```

```python
dados.head(10)
```

```python
casa = ['Casa', 'Casa de Condominio', 'Casa de Vila']
```

```python
dados['Tipo Agregado'] = dados[ 'Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

```python
dados
```

##Identificando e removendo Outliers

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure',figsize = (14, 6 ))
```

```python
dados = pd.read_csv('aluguel_residencial_2.csv', sep = ';')
```

```python
dados.boxplot(['Valor'])
```

```python
dados[dados['Valor']>= 500000]# Estamos redefinindo dados para valores até 500000 ou seja, removendo dados acima deste valor, retirando os outliers. 
```

```python
valor = dados['Valor']
```

```python
Q1 = valor.quantile(.25)
Q3 = valor.quantile(.75)
IIQ = Q3 - Q1
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ
```

```python
selecao = (valor >= limite_inferior) & (valor <= limite_superior)
dados_new = dados[selecao]
```

```python
dados_new.boxplot(['Valor'])
```

```python
dados.hist(['Valor'])
dados_new.hist(['Valor'])
```

```python
dados.boxplot(['Valor'], by = ['Tipo'])
```

```python
grupo_tipo = dados.groupby('Tipo')['Valor']
```

```python
type(grupo_tipo)
```

```python
grupo_tipo.groups
```

```python
Q1 = grupo_tipo.quantile(.25)
Q3 = grupo_tipo.quantile(.75)
IIQ = Q3 - Q1
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ
```

```python
Q1
```

```python
Q3
```

```python
IIQ
```

```python
limite_inferior
```

```python
limite_superior
```

```python
#Para selecionar o limite_superior de um tipo (Apartamento, casa ou casa de vila) específico 
limite_superior['Casa']
```

```python
dados_new = pd.DataFrame()
for tipo in grupo_tipo.groups.keys():
  eh_tipo = dados['Tipo'] == tipo
  eh_dentro_limite = (dados['Valor']>= limite_inferior[tipo]) & (dados['Valor'] <= limite_superior[tipo])
  selecaio = eh_tipo & eh_dentro_limite
  dados_selecao = dados[selecao]
  dados_new = pd.concat([dados_new, dados_selecao])
```

```python
dados_new.boxplot(['Valor'], by = ['Tipo'])
```

```python
dados_new.to_csv('aluguel_residencial_2_sem_outliers.csv', sep = ';', index = False)
```
