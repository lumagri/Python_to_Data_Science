#Métodos de Interpolação

```python
import pandas as pd
```

```python
data = [0.5, None, None, 0.53, 0.54, None, 0.7, None, 0.7]
s = pd.Series(data)
s
```

```python
s.fillna(0) # Preenche os valores nulos com o valor determinado na função fillna(n)
```

```python
s.fillna(method = 'ffill') # Este método de interpolação são bem válidos quando precisa se preencher valores de acordo com a passagem de tempo. Quando não a função encontra um valor nulo, ela preenche com o ultimo valor válido a cima do valor nulo.
```

```python
s.fillna(method = 'bfill')# Faz a mesma coisa que o ffill mas preenche com o ultimo valor válido que esteja abaixo b de below.
```

```python
s.fillna(s.mean()) # preenche com a média dos valores válidos. 
```

```python
s.fillna(method = 'ffill', limit = 1)# Faz o ffill com o número do limíte estabelecido. Neste caso aqui apenas 1 valor nulo será preenchido com o valor acima dele
```

```python
s.fillna(method = 'bfill', limit = 1)
```
