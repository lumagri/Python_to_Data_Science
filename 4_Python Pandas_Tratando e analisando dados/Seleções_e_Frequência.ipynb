##Tarefas 

* Selecionar os imóveis classificados como tipo "Apartamento"
* Selecionar os imóveis classificados com tipo "Casa", "Casa de Condomínio", "Casa de Vila".
* Selecionar imóveis com área entre 60 e 100 metros quadrados, incluindo os limites.
* Selecionar os imóveis que tenham pelo menos 4 quartos e aluguel menor que R$2.000,00



```python
import pandas as pd
```

```python
dados = pd.read_csv('aluguel_residencial.csv', sep = ';')
```

```python
dados.head(10)
```

```python
dados.shape[0] # Apresenta o número de linhas.
```

```python
dados.shape # Apresenta o número de linhas e colunas. 
```

```python
# Nº de imóveis classificados como tipo "Apartamento"
selecao = dados['Tipo'] == 'Apartamento'
selecao
n1 = dados[selecao].shape[0]
n1
```

```python
# Selecionar os imóveis classificados com tipo "Casa", "Casa de Condomínio", "Casa de Vila".
selecao = (dados['Tipo'] == 'Casa') |(dados['Tipo'] == 'Casa de Condomínio') |(dados['Tipo'] == 'Casa de Vila')
n2 = dados[selecao].shape[0] # Se retirar o shape aqui ele mostra a tabela com os elementos. Mas queremos saber a quantidade desses tipos de imóveis por isso colocamos o shape. 
n2
```

```python
# Selecionar imóveis com área entre 60 e 100 metros quadrados, incluindo os limites.
#60 <= Area <= 100 

selecao = (dados['Area']>=60) & (dados['Area'] <= 100)
n3 = dados[selecao].shape[0]
n3
```

```python
# Selecionar os imóveis que tenham pelo menos 4 quartos e aluguel menor que R$2.000,00
selecao = (dados['Quartos'] >= 4) & (dados['Valor'] < 2000)
n4 = dados[selecao].shape[0]
n4
```

```python
print("Nº de imóveis classificados como tipo 'Apartamento' => {}".format(n1))
print("Nº deimóveis classificados com tipo 'Casa', 'Casa de Condomínio', 'Casa de Vila'. => {}".format(n2))
print("Nº deimóveis com área entre 60 e 100 metros quadrados, incluindo os limites. => {}".format(n3))
print("Nº deimóveis que tenham pelo menos 4 quartos e aluguel menor que R$2.000,00 => {}".format(n4))
```
