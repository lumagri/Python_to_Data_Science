## Imóveis Residênciais

```python
import pandas as pd
```

```python
dados = pd.read_csv("aluguel.csv", sep = ';')
```

```python
dados.head(10)
```

Precisamos apenas trabalhar com a variável "Tipo" e não com a tabela completa. Então a partir daqui, todos os imóveis que não forem tipo residencial precisam ser retirados da lista. Queremos apenas os Tipo residencial. 

Para isso, valos listar os dados e cortar as informações que não são de imóveis "Tipo" Residêncial usando o drop_duplicates() 

```python
list(dados['Tipo'].drop_duplicates()) # Copio toda esta lista e crio a variável residencial  = [] mas retiro todos os itens que não forem residencial. 
```

```python
residencial = ['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
 'Casa de Vila']
```

```python
residencial
```

```python
dados.head(10)
```

```python
selecao = dados['Tipo'].isin(residencial)
selecao
```

```python
dados_residencial = dados[selecao]
```

```python
dados_residencial
```

```python
list(dados_residencial['Tipo'].drop_duplicates())
```

```python
dados_residencial.shape[0] # Shape verifica o tamanho do DataFrame
```

```python
dados.shape[0] # Shape verifica o tamanho do DataFrame
```

```python
dados_residencial.index = range(dados_residencial.shape[0]) # Para reconstruir o index usa-se o .index e o range como apresentado aqui. 
```
